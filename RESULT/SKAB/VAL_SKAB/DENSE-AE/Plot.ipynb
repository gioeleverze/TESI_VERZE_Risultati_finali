```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iterator=1
somma=[]
for i in range(33):
    file=np.load(f'{iterator}.npy')

    plt.figure(figsize=(4,5))
    plt.boxplot(file)
    plt.grid()

    
    somma.append((np.max(file)-np.mean(file))/np.std(file))
    iterator=iterator+1

print(np.array(somma).mean())
```
