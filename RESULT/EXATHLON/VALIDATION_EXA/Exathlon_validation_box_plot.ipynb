```python
file=np.load('DENSE_AE.npy')
print(f'mean:{np.mean(file)}')
print(f'std:{np.std(file)}')
print(f'max:{np.max(file)}')
plt.figure(figsize=(8,15))
plt.boxplot(file)
plt.grid()

print((np.max(file)-np.mean(file))/np.std(file))
score=file
t=0.5
 #IQR
q1, q3 = np.percentile(score, 25), np.percentile(score, 75)
iqr = q3 - q1
IQR = q3 + t* iqr

#MAD
median = np.median(score)
mad = 1.4826 * np.median(np.abs(score - median))
MAD = median + t * mad

#STD
mean, std = np.mean(score), np.std(score)
STD = mean + t * std

print(f'IQR:{IQR}')
print(f'STD:{STD}')
print(f'MAD:{MAD}')
```
